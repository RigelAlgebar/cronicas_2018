```julia
function test(x, y)
    if x < y
        print(x)
        print(" es menor que ")
        println(y)
    elseif x > y
        print(x)
        print(" es mayor que ")
        println(y)
    else
        print(x)
        print(" es igual que ")
        println(y)
    end
end
```

```julia
test(1,2)
test(2,1)
test(2,2)
```
