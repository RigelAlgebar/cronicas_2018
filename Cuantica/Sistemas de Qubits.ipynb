<h1 style="color:#0027d6;">Estados de un Bit Cuantico</h1>

Los qubits elementales $\mid0\rangle$ y $\mid1\rangle$ forman lo que en algebra lineal como una 'base' de un espacio vectorial en dos dimensiones, es decir, para formar cualquier otro elemento en 2D basta con especificar las amplitudes de los elementos base. En mecanica cuantica, decimos que cualquier otro estado cuantico formado por una combinacion lineal de los elementos base, es un estado en superposion cuantica de dichos estados.

<h1 style="color:#0027d6;">Estados de dos Bits Cuanticos</h1>

Los qubits que hemos discutido y construido hasta ahora son qubits simples, por ejemplo la siguiente combinacion lineal:

$$ \left| \psi_n \right\rangle = \alpha_n \left| 0 \right\rangle + \beta_n \left| 1 \right\rangle = \begin{pmatrix} \alpha_n \\ \beta_n \end{pmatrix} \: $$

Es la forma en que construyes 'n' qubits, tomando n distintas amplitudes para tus qubits elementales. Esto se conoce como superposicion de estados cuanticos. Pero existe una forma aun mas poderosa de combinar los elementos de informacion cuantica. Albert Einsten, Boris Podolsky y Nathan Rosen descubrieron en 1935 lo que conocemos hoy como: entrelazamiento cuantico. El paper en el que publicaron este resultado es por cierto el paper de la fisica mas citado de la hostoria. Cuando dos estados caunticos se encuentran entrelazados, obtenemos un nuevo sistema en el que no puedes describir de forma independiente cada una de las partes de las que fue creado.    

Cual es la diferencia en superposicion y entrelazamiento? 

La superposicion puede ocurrir para distintos estados cuanticos de un mismo sistema o distintos sistemas, mientras que el entrelazamiento esta limitado a mas de un sistema. Veamos esto un ejemplo fisico: un electron con spin arriba y spin abajo. 

Tu puedes crear superposicion de los dos spines del electron 'A', y al medir su spin obtendras una probabilidad de medir 'arriba' o 'abajo' dependiendo de las amplitudes que usaste para crear la superposicion. 

En el entrelazamiento, necesitas una clara distincion entre dos sistemas fisicos. Por ejemplo un electron 'A' y un electron 'B'. Si cada uno de ellos esta en superposicion de ciertos estados cuanticos, puedes medirlos sin problema, pero si estos estuvieran entrelazados, no puedes medir 'A' sin afectar 'B'. Al entrelazarlos, existe un intercambio de informacion entre ellos porque juntos han formado un nuevo sistema!
Ya no puedes ver al electron 'A' solo como el electron 'A', y lo mismo para el 'B'; ahora tienes un nuevo sistema entrelazado compuesto de 'A' y 'B'. 

El entrelazamiento cuantico es fundamental para la informacion y computacion cuantica. Entre sus aplicaciones se encuentran:

- Teletransportacion cuantica
- Criptografia cuantica
- Computacion cuantica

Y la lista sigue creciendo. Pero antes de discutir estas aplicaciones y puertas cuanticas, necesitamos discutir y estudiar el entrelazamiento. 

Para empezar, vamos a necesita una nueva operacion matematica llamada 'producto tensorial', este producto nos ayuda a construir estados de mas de un qubit. Veamos como se define para dos qubits $\mid\psi_1\rangle$ y $\mid\psi_2\rangle$.

$$ \mid\psi_1\rangle\otimes\mid\psi_2\rangle = \alpha_1\alpha_2\mid00\rangle + \alpha_1\beta_2\mid01\rangle + \beta_1\alpha_2\mid10\rangle + \beta_1\beta_2\mid11\rangle = \begin{pmatrix}\alpha_1\alpha_2 \\ \alpha_1\beta_2 \\ \beta_1\alpha_2 \\ \beta_1\beta_2 \end{pmatrix}$$

Donde:

$$ \mid00\rangle = \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix}, \quad \mid01\rangle = \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}, \quad \mid10\rangle = \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix}, \quad \mid11\rangle = \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix} $$

Representan los qubits elementales (base) del espacio de estados de dos qubits.
